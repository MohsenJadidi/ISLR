# Part A

```
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X_train, Y_train = make_moons(n_samples=1000, shuffle=True, noise=0.35, random_state=None)
X_test, Y_test = make_moons(n_samples=500, shuffle=True, noise=0.35, random_state=None)
```

# Part B

```
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

k_max = 500
errTrain = []
errTest = []
for n in range(k_max):
    classifier = KNeighborsClassifier(n_neighbors=n+1, weights='uniform')
    classifier.fit(X_train, Y_train)
    errTrain.append(1-classifier.score(X_train, Y_train))
    errTest.append(1-classifier.score(X_test, Y_test))

# Error respect to k
fig = plt.figure(num=1)
ax = fig.add_subplot(1, 1, 1)
k = [i+1 for i in range(k_max)]
line1, line2 = ax.plot(k, errTrain, 'b', k, errTest, 'r')
ax.set_xlabel(r'$k$')
ax.set_ylabel('Error')
ax.legend(['Train', 'Test'])
ax.set_xlim([-10,k_max])
ax.grid(b= True, which='both')

# Error respect to 1/k
fig = plt.figure(num=2)
ax = fig.add_subplot(1, 1, 1)
k = [1/(i+1) for i in range(k_max)]
k = [i for i in reversed(k)]
line1, line2 = ax.plot(k, [i for i in reversed(errTrain)] , 'b', 
                       k, [i for i in reversed(errTest)], 'r')
ax.set_xlabel(r'$ \frac{1}{k} $')
ax.set_ylabel('Error')
ax.legend(['Train', 'Test'])
ax.set_xscale('log')
ax.grid(b= True, which='both')
plt.tight_layout()
```
